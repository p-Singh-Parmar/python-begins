```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt                  #for data visualization

import seaborn as sns                            #advance data visualization library
```

```python
student = {"bob":87, "matt":34, "jack":11}
names= list([10,20,30])
values= list(student.values())
plt.bar(names, values)
plt.title("bar plot")
plt.xlabel("names")
plt.ylabel("values")
plt.grid(True)
plt.show()
```

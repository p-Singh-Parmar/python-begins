```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,5)
x
```

```python
y = x*2
y1= x*3
```

```python
y
```

```python
plt.plot(x,y,color='g', linestyle=':', linewidth=2)
plt.plot(x,y1,color='y', linestyle='-', linewidth=4)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
#margin labels can't be seen as black font on black background
```

```python
plt.subplot(1,2,1) #1 row, 2 columns, 1st column
plt.plot(x,y,color='g', linestyle=':', linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y1,color='y', linestyle='-', linewidth=4)
plt.title("Line Plot")

plt.show()
```

```python
student = {"bob":87, "matt":34, "jack":11}
```

```python
names= list(student.keys())
values= list(student.values())
```

```python
plt.bar(names, values)
plt.title("bar plot")
plt.xlabel("names")
plt.ylabel("values")
plt.grid(True)
plt.show()
```

```python
 plt.barh(names, values,color='g')
plt.title("bar plot")
plt.xlabel("names")
plt.ylabel("values")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,color='y',marker="*",s=100)
plt.subplot(1,2,2)
plt.scatter(x,y1,color='r',marker=".",s=100)
plt.show()
```
